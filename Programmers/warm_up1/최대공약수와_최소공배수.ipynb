- 최대공약수(GCD)
    - 두 수 혹은 여러 수의 공통인 약수 중, 최대인 것.
    - 최대공약수가 1이면 두 수는 서로소(coprime) 관계
    
    image.png

    ```
    8의 약수 - 1,2,4,8
    10의 약수 - 1,2,5,10
    8과 10의 최대공약수:2
    ```
- 최소공배수(LCM)
    - 두 수, 혹은 여러 수의 공통인 배수 중, 가장 작은 수
    
    image.png
    ```
    10의 배수 - 10,20,30,40...
    12의 배수 - 12,24,36,48,60...
    10과 12의 최소공배수:60
    ```

- 기본 알고리즘

````python
a=10
b=12

#최대공약수
for i in range(min(a,b),0,-1):
    if a%i==0 and b%i==0:
        print(i)
        break

#최소공배수
for i in range(max(a,b),(a*b)+1):
    if i%a == 0 and i%b ==0:
        print(i)
        break
````

```
[문제 설명]
두 수를 입력받아 두 수의 최대공약수와 최소공배수를 반환하는 함수, solution을 완성해 보세요. 
배열의 맨 앞에 최대공약수, 그다음 최소공배수를 넣어 반환하면 됩니다. 
예를 들어 두 수 3, 12의 최대공약수는 3, 최소공배수는 12이므로 
solution(3, 12)는 [3, 12]를 반환해야 합니다.

입출력 예
n	m	 return
3	12	[3, 12]
2	5	 [1, 10]
```

````python
def solution(n, m):
    answer = []
    for i in range(min(n,m),0,-1):
        if n%i==0 and m%i==0:
            answer.append(i)
            break
    for i in range(max(n,m),(n*m)+1):
        if i%n==0 and i%m==0:
            answer.append(i)
            break
    return answer
solution(3,12)
````
